```python
import subprocess
import os
from pathlib import Path
```

```python
path = Path().cwd().parent.joinpath('data/video_import')
filenames = os.listdir(path)
```

```python
filenames
```

```python
def video_rename():
    for filename in filenames:
        os.rename(os.path.join(path, filename), os.path.join(path, filename.replace(' ', '_').replace('(', '').replace(')', '')))
```

```python
video_rename()
```

```python
def convert_video_to_audio(video_file_path, audio_file_path):
    command = "ffmpeg -i f{video_file_path} -vn f{audio_file_path}"
    subprocess.call(command, shell=True)
```

```python
for file in os.getcwd():
    convert_video_to_audio(file, f"{file}.mp3")
```
